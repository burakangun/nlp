```python
# Metin Sınıflandırma

import pandas as pd
import on_islem
```

```python
df = pd.read_csv("data/metin_siniflandirma-230518-122016.csv",index_col=0)
df["Text2"] = df["Text"].apply(on_islem.pre_processing)
df["Text2"] = df["Text2"].apply(on_islem.remove_space)
df["Text3"] = [' '.join(wrd for wrd in x) for x in df["Text2"]]

df_index = df[df["Text2"].str[0].isnull()].index
df = df.drop(df_index)
df = df.reset_index()
del df["index"]
df.head()
```
